Segundo Ejemplo - Datos de olas en diferentes puertos
Queremos mostrar en un mapa 3 puertos de un Estado y cuando haga clic en un puerto me muestre data como viento, la altitud de olas, viento, lluvia, etc.
VAMOS A USAR UN ARCHIVO JSON

```python
import json

import requests

url = (
    "https://raw.githubusercontent.com/python-visualization/folium/main/examples/data"
)
vis1 = json.loads(requests.get(f"{url}/vis1.json").text)
vis2 = json.loads(requests.get(f"{url}/vis2.json").text)
vis3 = json.loads(requests.get(f"{url}/vis3.json").text)
```

```python
vis3
```

```python
import folium
```

```python
# GENERAR EL MAPA Y POSICIONARSE EN UN PUNTO DE INTERÉS

#m = folium.Map(location=[46.3014, -123.7390], zoom_start=7, tiles="Stamen Terrain", control_scale=True)
m = folium.Map(location=[46.3014, -123.7390], zoom_start=7, tiles="cartodb positron", control_scale=True)
m
```

```python
# CARGO LOS PUERTOS, ACA LOS HICIMOS A MANO PORQUE ERAN 3 

folium.Marker(
   
    location=[47.3489, -124.708],
    popup=folium.Popup(max_width=450).add_child( #POP UP ES EL METODO QUE MUESTRA L AINFO INTERACTIVA
        folium.Vega(vis1, width=450, height=250) #VEGA le dice cual json quiero que gráfique
    ),
).add_to(m)

folium.Marker(
    location=[44.639, -124.5339],
    popup=folium.Popup(max_width=450).add_child(
        folium.Vega(vis2, width=450, height=250)
    ),
).add_to(m)

folium.Marker(
    location=[46.216, -124.1280],
    popup=folium.Popup(max_width=450).add_child(
        folium.Vega(vis3, width=450, height=250)
    ),
).add_to(m) # AGREGARLOS TODOS AL MAPÀ


m
```

```python
#EXPORTAMOS NUESTRO MAPA EN UN HTML QUE SERÍA UN ENTREGABLE.
#el archivo se guarda en la ruta donde estas almacenando los notebooks, en este ejercicio por defecto en mi pc local.
m.save("mi_mapa_interactivo.html")
```
